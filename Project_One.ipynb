```python
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Import the CSV GDP to Pandas DataFrames
file_path_2 = Path("COVID19_Impact_on_Global_GDP.csv")
GDP_data = pd.read_csv(file_path_2, index_col='Year')
GDP_data.head()
```

```python
# Import the CSV Labour Market to Pandas DataFrames
file_path_3 = Path('Labour_Market_stats.csv')
Labour_Market_stats = pd.read_csv(file_path_3, index_col='Time')
Labour_Market_stats.head()
```

```python
# Import the CSV Labour Market to Pandas DataFrames
file_path_4 = Path("COVID19_Impact_per_Measure.csv")
Impact_per_Measure = pd.read_csv(file_path_4, index_col="Years")
Impact_per_Measure.head()
```

```python
# Using hvplot to create an interactive bar chart of the GDP per Capita of each country including a dropdown selector for the neighbourhood
Per_Capita = ['GDP per Capita']
GDP_data.hvplot.bar('Year',
                   Per_Capita,groupby='Country',
                   rot=90,
                   stacked=False,
                   xlabel='Years',
                   ylabel='GDP per Capita: in U.S. dollars',
                   width=700, 
                   height=400)
```

```python
# Import the CSV GDP to Pandas DataFrames
file_path_2 = Path("COVID19_Impact_on_Global_GDP.csv")
GDP_data2 = pd.read_csv(file_path_2)
GDP_data2.head()
```

```python
# Using hvplot to create an interactive line chart of the GDP per Country
GDP_data2.hvplot.line(x='Year',y='GDP',groupby='Country')
```

```python
# Using hvplot to create an interactive bar chart of the GDP Growth Rate per Country with a dropdown selector for each measure
Impact_per_Measure.hvplot.bar('Years', ['China','Japan','Brazil','Canada','United States','United Kingdom','Italy','India','France','Germany'],
                   groupby='Measures',
                   rot=90,
                   stacked=False,
                   xlabel='Country',
                   ylabel='Percentage GDP Growth Rate',
                   width=1400, 
                   height=700)
```

```python
#Line graph Showing Employment Rate per country.
file_path_3 = Path('Labour_Market_stats.csv')
Labour_Market_stats = pd.read_csv(file_path_3, index_col='Time')
Labour_Market_stats.drop(columns=['France','Germany','Italy','Japan','United Kingdom','United States'],inplace=True)
Labour_Market_stats.head()
```

```python
#Covid-19 effect on Global Employment
Labour_Market_stats.hvplot(x='Time', y=['Canada', 'G7', 'OECD_Total','Euro_Area_19','European_Union_27'],
             value_label='Employment Rate (Percentage of Population)',
             width=800,
             height=500,
             title='Covid-19 effect on Global Employment')
```
